```python
import sys

def solve():
    # Dictionary to store stats: {player_name: [5-set-won, 3-set-won, sets-won, games-won, sets-lost, games-lost]}
    stats = {}

    def get_player(name):
        if name not in stats:
            stats[name] = [0, 0, 0, 0, 0, 0]
        return stats[name]

    for line in sys.stdin:
        line = line.strip()
        if line == "EOF" or not line:
            break

        # Split into Player1, Player2, and the Sets
        try:
            players_part, scores_part = line.split(':')
            p1_name, p2_name = players_part.split('-')
            sets = scores_part.split(',')
        except ValueError:
            continue

        p1 = get_player(p1_name)
        p2 = get_player(p2_name)

        p1_sets_won = 0
        p2_sets_won = 0

        for s in sets:
            g1, g2 = map(int, s.split('-'))
            p1[3] += g1  # P1 games won
            p1[5] += g2  # P1 games lost
            p2[3] += g2  # P2 games won
            p2[5] += g1  # P2 games lost

            if g1 > g2:
                p1_sets_won += 1
            else:
                p2_sets_won += 1

        p1[2] += p1_sets_won
        p1[4] += p2_sets_won
        p2[2] += p2_sets_won
        p2[4] += p1_sets_won

        # Determine match winner and match type
        total_sets = p1_sets_won + p2_sets_won
        is_five_set_match = total_sets > 3 or (p1_sets_won == 3 or p2_sets_won == 3)

        # Note: In tennis, 2 sets won usually means best-of-3, 3 sets won means best-of-5.
        if p1_sets_won > p2_sets_won:
            if p1_sets_won >= 3: p1[0] += 1
            else: p1[1] += 1
        else:
            if p2_sets_won >= 3: p2[0] += 1
            else: p2[1] += 1

    # Sorting logic:
    # Python's sort is stable. For descending order with some ascending criteria,
    # we can negate the values of ascending criteria (5 and 6).
    sorted_players = sorted(stats.items(),
                            key=lambda x: (x[1][0], x[1][1], x[1][2], x[1][3], -x[1][4], -x[1][5]),
                            reverse=True)

    for name, s in sorted_players:
        print(f"{name} {s[0]} {s[1]} {s[2]} {s[3]} {s[4]} {s[5]}")

if __name__ == "__main__":
    solve()
```
