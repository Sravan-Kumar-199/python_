Funtions:

```python
def fun_name(a,b):
   print("addition:",a+b)
   print("subtraction:",a-b)
   print("multiplication:",a*b)
   print("division:",a/b)
fun_name(50,20)
```

```python
def fun_name(a,b,c):
  print("addition:",a+b+c)
  print("subtraction:",a-b-c)
  print("multiplication:",a*b*c)
  print("division:",a/b/c)
fun_name(150,70,20)
```

```python
def fun_name(a,b,c,d):
  print("addition:",a+b+c+d)
  print("subtraction:",a-b-c-d)
  print("multiplication:",a*b*c*d)
  print("division:",a/b/c/d)
fun_name(100,200,300,400)
```

Write a program to find bill amount of a customer using functions:

```python
def callbill(incoming,outgoing):
  bill=incoming*outgoing
  print("bill amount:",bill)
callbill(100,200)
```

```python
def mandi(items,quantity):
  total=items+quantity
  print("total:",total)
mandi(15,35)
```

1.Local Variables-

```python
def fun_name():
  a=50
  print(a)
fun_name()
```

```python
def fun_name():
  a=78
  b=80
  print(a+b)
fun_name()
```

```python
def fun_name():
  a=15
  b=40
  print(a*b)
fun_name()
```

```python
def fun_name():
  a=78
  b=70
  c=90
  print(a+b+c)
fun_name()
```

```python
def fun_name():
  a=15
  b=55
  c=65
  print(a+b+c)
  print(a-b-c)
  print(a*b*c)
  print(a/b/c)
fun_name()
```
