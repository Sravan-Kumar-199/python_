Logical operators ___ This operator allows you to combine multiple comparisons to create  complex decision making rules...

```python
Operator Description : and - Returns True if both statements are true ,
                        or - Returns True if one of the statements is true ,
                        not - Reverse the result ...
```

AND operator example -- Use AND when multiple conditions must be met simultaneously...

```python
# Scenario: Checking if a student is eligible  for a scholarship
# Criteria: Grade must be > 90 AND attendence must be > 95%

grade = 92
attendence = 96

if grade > 90 and attendence > 95:
  print("Scholarship Eligibility: Confirmed")
else:
  print("Scholarship Eligibility: Denied")
```

OR operator example -- Use OR when you only need one of  the several conditions to be true...

```python
#  Scenario: Checking for weekend or holiday discount
day = "Saturday"

if day == "Saturday" or day == "Sunday" or day == "Holiday":
  print("Weekend or Holiday Discount: Yes")
else:
  print("Weekend or Holiday Discount: No")
```

NOT operator example -- It is used to  flip a Boolen value.it's often used to check if a condition is not met...

```python
# Scenario: Checking if a system is NOT locked
is_locked = False

if not is_locked:
  print("System is unlocked")
else:
  print("System is locked")
```
