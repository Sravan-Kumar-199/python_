Relation Operators-
Relation operators are berad and butter of logic in Pyton.They allow you to compare two values and always return a Boolen result;either True or False...

Operator And Meaning ___ == Equal to , != Not Equal to , > Greater than , < Less than , >= Greater than or equal to , <= Less than or equal to ...

Example ...

```python
a=10
b=20

print(f"Is {a} equal to {b}?",a==b)
print(f"Is {a} not equal to {b}?",a!=b)
print(f"Is {a} greater than {b}?",a>b)
print(f"Is {a} less than {b}?",a<b)
```
